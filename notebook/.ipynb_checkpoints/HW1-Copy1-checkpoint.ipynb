```python
class0count=0
class1count=0
for i in range(0,len(df['class'])):
    if df.loc[i]['class']==1.0:
        if class1count <140:
            df_train=df_train.append(df.loc[i],ignore_index=True)
            class1count+=1
        else:
            df_test=df_test.append(df.loc[i],ignore_index=True)
    else:
        if class0count <70:
            df_train=df_train.append(df.loc[i],ignore_index=True)
            class0count+=1
        else:
            df_test=df_test.append(df.loc[i],ignore_index=True)

df_test#testing dataset created
```

### ii. Test all the data in the test database with k nearest neighbors. Take decisions by majority polling. Plot train and test errors in terms of k fork âˆˆ {208, 205, . . . , 7, 4, 1, } (in reverse order). You are welcome to use smaller increments of k. Which kâˆ—is the most suitable k among those values? Calculate the confusion matrix, true positive rate, true negative rate, precision,and F1-score when k = kâˆ—.



```python
train_errors={}#dictionary which has the training error for k values
test_errors={}#dictionary which has the testing error for k values
for i in range(1,209,3):
    knn=KNeighborsClassifier(n_neighbors=i,p=2)
    knn.fit(df_x_train,df_y_train)
    train_errors[i]=1-knn.score(df_x_train,df_y_train)
    test_errors[i]=1-knn.score(df_x_test,df_y_test)
```

```python
sorted_train_dict=sorted(train_errors.items(),reverse=True)
sorted_test_dict=sorted(test_errors.items(),reverse=True)
```

```python
#plotting the training error and test error rates

k_values_train,error_values_train=zip(*sorted_train_dict)
k_values_test,error_values_test=zip(*sorted_test_dict)

plt.title('train and test errors w.r.t K')
plt.plot(k_values_train,error_values_train,label="Training error")
plt.plot(k_values_test,error_values_test,label="Testing error")
plt.legend()
plt.xlabel('K-values')
plt.ylabel('Error')
plt.axis([max(max(k_values_train),max(k_values_test)),min(min(k_values_train),min(k_values_test)),
          min(min(error_values_train),min(error_values_test)),
          max(max(error_values_train),max(error_values_test))])
# fig,(ax1,ax2)=plt.subplots(1,2)
# ax1.plot(k_values_train,error_values_train,label="Training error")
# ax2.plot(k_values_test,error_values_test,label="Testing error")
# ax1.invert_xaxis()
# ax2.invert_yaxis
# fig.tight_layout()
plt.show()
```

```python
max(max(error_values_train),max(error_values_test))
```

```python
#choosing k when lowest error occurs in testing set

min_error=min(test_errors.values())
opt_kvalues=[]
for key in test_errors:
    if(test_errors[key]==min_error):
        opt_kvalues.append(key)

opt_kvalues

#hence the optimum k value is 4
```

```python
#plotting confusion matrix when k=4

knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(df_x_train,df_y_train)
predicted_values=knn.predict(df_x_test)
print("confusion matrix fo k=4")
print('\n')
print(confusion_matrix(df_y_test,predicted_values))
```

```python
#finding precision and f1-score
print(classification_report(df_y_test,predicted_values))
```

```python
conf_matrix=confusion_matrix(df_y_test,predicted_values)
tpr=conf_matrix[1][1]/(conf_matrix[1][1]+conf_matrix[1][0])
tnr=conf_matrix[0][0]/(conf_matrix[0][0]+conf_matrix[0][1])
print("True Positive Rate :",tpr)
print("True Negative Rate :",tnr)
```

## (d) Replace the Euclidean metric with the following metrics and test them. Summarize the test errors (i.e., when k = kâˆ—) in a table. Use all of your training data and select the best k when {1, 6, 11, . . . , 196}.
### i. Minkowski Distance:
### A. which becomes Manhattan Distance with p = 1.

```python
test_errors_manhattan={}
for i in range(1,197,5):
    knn=KNeighborsClassifier(n_neighbors=i,p=1)
    knn.fit(df_x_train,df_y_train)
    test_errors_manhattan[i]=1-knn.score(df_x_test,df_y_test)
min_error=min(test_errors_manhattan.values())
opt_kvalues=[]
for key in test_errors_manhattan:
    if(test_errors_manhattan[key]==min_error):
        opt_kvalues.append(key)
print(opt_kvalues,min_error)
opt_kvalueman=opt_kvalues[0]
min_error_man=min_error
```

### B. with log10(p) âˆˆ {0.1, 0.2, 0.3, . . . , 1}. In this case, use the kâˆ— you found for the Manhattan distance in 1(d)iA. What is the best log10(p)?

```python
test_errors_log={}
for i in np.arange(0.1,1.1,0.1):
    knn=KNeighborsClassifier(n_neighbors=opt_kvalueman,p=pow(10,i))
    knn.fit(df_x_train,df_y_train)
    test_errors_log[i]=1-knn.score(df_x_test,df_y_test)
min_error=min(test_errors_log.values())
opt_logvalues=[]
for key in test_errors_log:
    if(test_errors_log[key]==min_error):
        opt_logvalues.append(key)
# print(opt_logvalues,min_error)
opt_logvalue=opt_logvalues[0]
min_error_log=min_error
print(opt_logvalue,min_error_log)
```
